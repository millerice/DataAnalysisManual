```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
df = pd.read_excel('./data/sales-funnel.xlsx')
```

```python
df
```

```python
# 生成透视表 自动聚合 默认按照平均值, 可自定义

pd.pivot_table(df, index=['Name', 'Rep', 'Manager'], aggfunc='sum')
```

```python
pd.pivot_table(df, index=['Manager', 'Rep'])
```

```python
# 控制展示的values
pd.pivot_table(df, index=['Manager', 'Rep'], values=['Price'], aggfunc='sum')
```

```python
# 控制展示的columns
pd.pivot_table(df, index=['Manager', 'Rep'], values=['Price', 'Quantity'], columns=['Product'],fill_value=0, aggfunc='sum')
```
