```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
df = pd.read_csv('./data/city_weather.csv')
```

```python
g = df.groupby('city')
```

```python
g
```

```python
g.describe()
```

```python
def foo(attr):
    return attr.max() - attr.min()
```

```python
# 自定义聚合函数
g.agg(foo)
```

```python
# 按照两个以上的值做group
g_new = df.groupby(['city', 'wind'])
```

```python
g_new
```

```python
# 取值
g_new.get_group(('BJ', 3))
```

```python
for (city_name, wind_name), group in g_new:
    print(city_name, wind_name)
    print(group)
```
